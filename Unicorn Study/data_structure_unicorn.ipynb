```python
companies['Date Joined'] = pd.to_datetime(companies['Date Joined'])
```

```python
companies.dtypes
```

```python
companies['Month Joined'] = companies['Date Joined'].dt.month_name()
companies.head(10)
```

```python
companies['Year Joined'] =  companies['Date Joined'].dt.year
companies['Years To Join'] = companies['Year Joined'] - companies['Year Founded']
companies.head(20)
```

```python
companies['Year Joined'].value_counts().sort_values(ascending=False)
```

```python
yr_of_intrst = companies[companies['Date Joined'].dt.year == 2021]
yr_of_intrst
```

```python

yr_of_intrst['wk_num'] = yr_of_intrst['Date Joined'].dt.strftime('%Y-W%V')
yr_of_intrst.head(10)
```

```python
wk_num = (yr_of_intrst[['wk_num','Company']].
             groupby(['wk_num'])
             .count()
             .reset_index()
             .rename(columns={"wk_num":"wk_2021", "Month Joined":"count"})
            
             )
wk_num.head(19)
```

```python
yr_2020 = companies[companies['Date Joined'].dt.year == 2020]
yr_2020
```

```python
yrs_2020_2021 = pd.concat([yr_of_intrst.drop(['wk_num'], axis = 1), yr_2020])
yrs_2020_2021.head(10)
```

```python
yrs_2020_2021['quarter'] = yrs_2020_2021['Date Joined'].dt.to_period('Q').dt.strftime('%Y-Q%q')
yrs_2020_2021
```

```python
valuation_time = (yrs_2020_2021[['quarter','Valuation']].
             groupby(['quarter'])
             .mean('Valuation')
             .reset_index()
             .rename(columns={"quarter":"yr_Q" ,"Valuation":"avg_valuation"})
            
             )
valuation_time.head(19)
```

```python

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']
```

```python
g = sns.boxplot(data = companies,
                x = 'Month Joined',
                y = 'Years To Join',
                hue = 'Month Joined',
                order= month_order,
                showfliers = False,
                );
plt.setp(g.get_xticklabels(), rotation=90)
g.set_title('Years to Join');


```

```python
avg_yrs_by_comp = (companies[['Year Founded','Years To Join']].
             groupby(['Year Founded'])
             .mean('Years To Join')
             .reset_index()
             .rename(columns={"Years To Join":"avg_num_yrs"})
            
             )
avg_yrs_by_comp
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x = avg_yrs_by_comp['Year Founded'], 
            y = avg_yrs_by_comp['avg_num_yrs'], 
            errorbar=('ci', False),
            hue = avg_yrs_by_comp['Year Founded'],
            )
plt.xlabel('Founding Yeards')
plt.ylabel('Average year')
plt.title("AVG Years Till Unicorn")
plt.xticks(rotation  = 45, fontsize = 8, horizontalalignment='right')
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(x = wk_num['wk_2021'], 
            y = wk_num['Company'], 
            errorbar=('ci', False),
            hue = wk_num['wk_2021'],
            )
plt.xlabel('Week Number')
plt.ylabel('Number Of Companies')
plt.title("Companies Founded in the Year 2021")
plt.xticks(rotation  = 45, fontsize = 8, horizontalalignment='right')
plt.show()
```
